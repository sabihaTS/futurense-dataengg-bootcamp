```python
import findspark
findspark.init()
```

```python
from pyspark import SparkContext
```

```python
sc=SparkContext("local")
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[1]") \
                    .appName('SparkByExamples.com') \
                    .getOrCreate()
```

```python
from pyspark.sql.functions import *
```

```python
rdd=sc.textFile("C:/Users/miles.MILE-BL-4409-LA/Downloads/modify1_weather.txt")
rdd1=rdd.map(lambda x:x.split(","))
```

```python
#Created DataFrame using RDD
df_1=spark.createDataFrame(rdd1)
```

```python
#Display weather data
df_1.show()
```

```python
df_1=df_1.withColumn("_6",col("_6").cast("float"))
df_1=df_1.withColumn("_7",col("_7").cast("float"))
```

```python
#Display Max Temperature
df_1.select(max("_6")).show()
```

```python
#Display min Temperature
df_1.select(min("_7")).show()
```

```python
#Display MonthWise max Temperature
df_1.groupBy(month("_2")).max("_6").show()
```

```python
#Display Monthwise Minimum Temperature
df_1.groupBy(month("_2")).min("_7").show()
```
